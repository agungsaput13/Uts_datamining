```python
x = dataGabungan.iloc[:,7: 76]
```

```python
x
```

```python
y = dataGabungan.iloc[:,83: 84]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.2, random_state = 23)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
```

```python
modelAgung = DecisionTreeClassifier(criterion='entropy', splitter = 'random')
modelAgung.fit(x_train,y_train)
```

```python
y_pred = modelAgung.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test,y_pred)
```

```python
accuracy
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig = plt.figure(figsize = (10, 7))
tree.plot_tree(modelAgung, feature_names = x.columns.values, class_names = np.array(['MITM ARP Spoofing','MQTT DoS Publish Flood','MQTT Malforme']), filled = True)
plt.show()
```

```python
import seaborn as lol
from sklearn import metrics
label = np.array([ 'MITM ARP Spoofing','MQTT DoS Publish Flood','MQTT Malforme'])
```

```python
import matplotlib.pyplot as plt
conf_matrix = metrics.confusion_matrix(y_test, y_pred)
plt.figure(figsize=(10, 10))
lol.heatmap(conf_matrix, annot=True, xticklabels=label, yticklabels=label)
plt.xlabel('Prediksi')
plt.ylabel('Fakta')
plt.show()
```
